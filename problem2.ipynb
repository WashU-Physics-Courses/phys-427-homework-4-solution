```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_euler = np.loadtxt('output-euler.csv', delimiter=',')
data_rk4 = np.loadtxt('output-rk.csv', delimiter=',')
plt.plot(data_euler[:,0], data_euler[:,1], label='Euler')
plt.plot(data_rk4[:,0], data_rk4[:,1], label='RK4')
ax = plt.gca()
ax.set_xlabel('x')
ax.set_ylabel('$\\theta$')
```
